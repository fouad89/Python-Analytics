```python
import requests
from bs4 import BeautifulSoup as soup
url='http://www.skysports.com/premier-league-fixtures'
response=requests.get(url)
page_soup=soup(response.content,'lxml')
#print(page_soup.prettify())
list_of_fixres=[]
side_1_list=[]
side_2_list=[]
fixr_sides=[]
fixr_body=page_soup.find_all('div',{'class':'fixres__body'})[0]
dates_of_fixr=fixr_body.find_all('h4',{'class':'fixres__header2'})
sides_of_fixr=fixr_body.find_all('div',{'class':'fixres__item'})
match_time=fixr_body.find_all('span',{'class':'matches__date'})
print(len(dates_of_fixr))
for side1 in range(len(sides_of_fixr)):
    side_1_team=sides_of_fixr[side1].find_all('span',{'class':'matches__item-col matches__participant matches__participant--side1'})[0].get_text()
    side_1_list.append(side_1_team.strip('\n'))
for side2 in range(len(sides_of_fixr)):
    side_2_team=sides_of_fixr[side2].find_all('span',{'class':'matches__item-col matches__participant matches__participant--side2'})[0].get_text()
    side_2_list.append(side_2_team.strip('\n'))
fixr_sides=[str(side_1_list[i])+' - '+str(side_2_list[i]) for i in range(len(side_1_list))]
'''for i in range(len(side_1_list)):
    print(side_1_list[i]+' - '+side_2_list[i])'''
print(len(fixr_sides))
for fixture in range(len(fixr_sides)):
    print (fixr_sides[fixture])
    print(match_time[fixture].get_text()+'\n')
```
